```python
#from selenium import webdriver
driver = webdriver.Chrome('/Users/donghunbyun/Desktop/workspace/jupyter/bigdata/chromedriver')

url = 'http://www.melon.com/chart/index.htm'
driver.get(url)

#from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

songs = soup.select('tbody > tr')  

for song in songs: 
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select ('div.ellipsis.rank02 > a')[0].text  
    print(title, singer, sep = ' | ')
```

```python
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)
songs = driver.find_elements_by_css_selector('tbody > tr')
for song in songs: 
    title = song.find_elements_by_css_selector('div.ellipsis.rank01 > span > a')[0].text
    singer = song.find_elements_by_css_selector('div.ellipsis.rank02 > a')[0].text  
    print(title, singer, sep = ' | ')
```

```python
song_data = []
rank = 1
songs = soup.select('tbody > tr')  
for song in songs: 
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    song_data.append(['Melon', rank, title, singer])
    rank = rank + 1
```

```python
song_data
```

```python
import pandas as pd
```

```python
columns = ['서비스', '순위', '타이틀', '가수']
```

```python
pd_data = pd.DataFrame(song_data, columns=columns)
pd_data.head()
```

```python
pd_data.to_excel('melon20210413.xlsx', index=False)
```
