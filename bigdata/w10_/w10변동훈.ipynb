```python
import unicodedata
```

```python
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    try:
        content = soup.select('div.C4VMK > span')[0].text
        content = unicodedata.normalize('NFC', content)
    except:
        content = ' '
        
    tags = re.findall(r'#[^\s#,\\]+', content)
    
    date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10]
    
    try:
        like = soup.select('div.Nm9Fw > a')[0].text[4:-1]   
    except:
        like = 0
        
    try: 
        place = soup.select('div.M30cS')[0].text
        place = unicodedata.normalize('NFC', place)
    except:
        place = ''
    data = [content, date, like, place, tags]
    
    return data
```

```python
from selenium import webdriver

driver = webdriver.Chrome("/Users/donghunbyun/Desktop/workspace/jupyter/bigdata/chromedriver")
```

```python
def youtube_searching(page):
    url = f"https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={page}"
    return url
```

```python
import re
```

```python
def transfer_subscriber(sub_cnt):
    replaced_subscriber = sub_cnt.replace('Îßå', '0000')
    replaced_subscriber = re.sub('[^0-9]','',replaced_subscriber)
    
    return replaced_subscriber
```

```python
import unicodedata
```

```python
def get_rank(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    ranks = soup.select('table > tbody> tr.aos-init')
    
    data = []
    
    for rank in ranks:
        title = rank.select('td.subject > h1 > a')[0].text.strip()
        title = unicodedata.normalize('NFC', title)
        category = rank.select('td.subject > h1 > p.category')[0].text.strip()
        category = unicodedata.normalize('NFC', category)
        subscriber = rank.select('td.subscriber_cnt')[0].text.strip()
        subscriber = unicodedata.normalize('NFC', subscriber)
        view = rank.select('td.view_cnt')[0].text.strip()
        view = unicodedata.normalize('NFC', view)
        video = rank.select('td.video_cnt')[0].text.strip()
        video = unicodedata.normalize('NFC', video)
        
        replaced_subscriber = transfer_subscriber(subscriber)
                
        data.append([title, category, subscriber, view, video, replaced_subscriber])
    
    return data
```

```python
result = []

for i in range(1, 11):
    url = youtube_searching(i)

    driver.get(url)
    
    data = get_rank(driver)
    
    result += data
```

```python
df = pd.DataFrame(result)
df.columns = ["title", "category", "subscriber", "view", "video", "replaced_subscriber"]
```
